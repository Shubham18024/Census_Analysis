```python
#Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats

# Read the CSV file
file_path = "C://Users//user//OneDrive//Desktop//2011 Data Education^J Population^J Religion^J Workers.csv"
population_data = pd.read_csv(file_path)
population_data
```

```python
# Calculating percentages
population_data['main_worker_percentage'] = (population_data['Main_Workers_Population'] / population_data['Total_Population']) * 100
population_data['marginal_worker_percentage'] = (population_data['Marginal_Workers_Population'] / population_data['Total_Population']) * 100

# Spearman correlation between Literacy_rates and main_worker_percentage
spearman_main, _ = stats.spearmanr(population_data['Literacy_rates'], population_data['main_worker_percentage'])
print(f"Spearman correlation (Main Worker Percentage): {spearman_main:.3f}")

# Spearman correlation between Literacy_rates and marginal_worker_percentage
spearman_marginal, _ = stats.spearmanr(population_data['Literacy_rates'], population_data['marginal_worker_percentage'])
print(f"Spearman correlation (Marginal Worker Percentage): {spearman_marginal:.3f}")

# Pearson correlation between Literacy_rates and main_worker_percentage
pearson_main, _ = stats.pearsonr(population_data['Literacy_rates'], population_data['main_worker_percentage'])
print(f"Pearson correlation (Main Worker Percentage): {pearson_main:.3f}")

# Pearson correlation between Literacy_rates and marginal_worker_percentage
pearson_marginal, _ = stats.pearsonr(population_data['Literacy_rates'], population_data['marginal_worker_percentage'])
print(f"Pearson correlation (Marginal Worker Percentage): {pearson_marginal:.3f}")

# Setting up plot area with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Plot 1: Literacy rate vs Main Worker Percentage
ax1.scatter(population_data['Literacy_rates'], population_data['main_worker_percentage'], color='blue', label='state', s=10)
ax1.set_facecolor('lightyellow')
ax1.set_title('Literacy Rate vs Main Worker Percentage')
ax1.set_xlabel('Literacy Rate')
ax1.set_ylabel('Main Worker Percentage (%)')
ax1.set_ylim(0, 100)
ax1.legend()

# Plot 2: Literacy rate vs Marginal Worker Percentage
ax2.scatter(population_data['Literacy_rates'], population_data['marginal_worker_percentage'], color='green', label='state', s=10)
ax2.set_facecolor('lightyellow')
ax2.set_title('Literacy Rate vs Marginal Worker Percentage')
ax2.set_xlabel('Literacy Rate')
ax2.set_ylabel('Marginal Worker Percentage (%)')
ax2.set_ylim(0, 100)
ax2.legend()

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()


```

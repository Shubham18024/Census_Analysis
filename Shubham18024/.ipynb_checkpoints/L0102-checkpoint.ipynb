```python
#Education level (1st -12th class and Graduate & Above) in each religion (2011)

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\saim\Just_one_for_all\Desktop\Census_Analysis\CSV files for 1991,2001,2011\2011 Data Education^J Population^J Religion^J Workers.csv")

df.columns
```

```python

columns = [
    'Hindu_Population ', 'Muslim_Population ', 'Christian_Population', 'Sikh_Population', 
    'Buddhist_Population ', 'Jain_Population ', 
    'Higher Secondary/Intermediate/Pre-University/Senior secondary_Males', 
    'Higher Secondary/Intermediate/Pre-University/Senior secondary_Females', 
    'Graduate & above_Males', 'Graduate & above_Females'
]

df_selected = df[columns]

df_selected.columns = [
    'Hindu_Population', 'Muslim_Population', 'Christian_Population', 'Sikh_Population', 
    'Buddhist_Population', 'Jain_Population', 
    'Higher_Secondary_Males', 'Higher_Secondary_Females', 
    'Graduate_Males', 'Graduate_Females'
]

# Calculating total higher secondary and graduate population for each religion
df_selected = df_selected.assign(
    Higher_Secondary_Total = df_selected['Higher_Secondary_Males'] + df_selected['Higher_Secondary_Females'],
    Graduate_Total = df_selected['Graduate_Males'] + df_selected['Graduate_Females']
)

total_population = {
    'Hindu': df_selected['Hindu_Population'].sum(),
    'Muslim': df_selected['Muslim_Population'].sum(),
    'Christian': df_selected['Christian_Population'].sum(),
    'Sikh': df_selected['Sikh_Population'].sum(),
    'Buddhist': df_selected['Buddhist_Population'].sum(),
    'Jain': df_selected['Jain_Population'].sum()
}

# Aggregating the total higher secondary and graduate population for each religion
total_higher_secondary = df_selected['Higher_Secondary_Total'].sum()
total_graduate = df_selected['Graduate_Total'].sum()

# Calculating higher secondary and graduate pass rates for each religion
higher_secondary_pass = {
    'Hindu': total_higher_secondary * total_population['Hindu'] / df['Total_Population'].sum(),
    'Muslim': total_higher_secondary * total_population['Muslim'] / df['Total_Population'].sum(),
    'Christian': total_higher_secondary * total_population['Christian'] / df['Total_Population'].sum(),
    'Sikh': total_higher_secondary * total_population['Sikh'] / df['Total_Population'].sum(),
    'Buddhist': total_higher_secondary * total_population['Buddhist'] / df['Total_Population'].sum(),
    'Jain': total_higher_secondary * total_population['Jain'] / df['Total_Population'].sum()
}

print(higher_secondary_pass)

graduate_pass = {
    'Hindu': total_graduate * total_population['Hindu'] / df['Total_Population'].sum(),
    'Muslim': total_graduate * total_population['Muslim'] / df['Total_Population'].sum(),
    'Christian': total_graduate * total_population['Christian'] / df['Total_Population'].sum(),
    'Sikh': total_graduate * total_population['Sikh'] / df['Total_Population'].sum(),
    'Buddhist': total_graduate * total_population['Buddhist'] / df['Total_Population'].sum(),
    'Jain': total_graduate * total_population['Jain'] / df['Total_Population'].sum()
}

religions = list(total_population.keys())

x = range(len(religions))

fig, ax = plt.subplots(figsize=(12, 6))

# Higher secondary pass rate
ax.bar(x, [higher_secondary_pass[religion] for religion in religions], width=0.4, label='Higher Secondary Pass', align='center')

# Graduate pass rate
ax.bar(x, [graduate_pass[religion] for religion in religions], width=0.4, label='Graduate Pass', align='edge')

ax.set_xlabel('Religions')
ax.set_ylabel('Number of People')
ax.set_title('Educational Attainment by Religion in India (2011)')
ax.set_xticks(x)
ax.set_xticklabels(religions)
ax.legend()

plt.show()
```

```python
higher_secondary_pass_percentage = {
    religion: (df_selected['Higher_Secondary_Total'].sum() / total_population[religion]) * 100
    for religion in total_population.keys()
}

graduate_pass_percentage = {
    religion: (df_selected['Graduate_Total'].sum() / total_population[religion]) * 100
    for religion in total_population.keys()
}

print(graduate_pass_percentage)

religions = list(total_population.keys())
x = range(len(religions))

fig, ax = plt.subplots(figsize=(12, 6))

# Higher secondary pass rate
ax.bar(x, [higher_secondary_pass_percentage[religion] for religion in religions], width=0.4, label='Higher Secondary Pass (%)', align='center')

# Graduate pass rate
ax.bar(x, [graduate_pass_percentage[religion] for religion in religions], width=0.4, label='Graduate Pass (%)', align='edge')

ax.set_xlabel('Religions')
ax.set_ylabel('Percentage of People')
ax.set_title('Educational Attainment by Religion in India (2011)')
ax.set_xticks(x)
ax.set_xticklabels(religions)
ax.legend()

plt.show()
```

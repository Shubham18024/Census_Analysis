```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
df = pd.read_csv('CENSUS1991.csv')

# Assuming the CSV has columns 'States' and 'Literacy Rate'
# Sort data by literacy rate for better visualization
df = df.sort_values(by='Literacy_Rate%', ascending=False)

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Create bar plot using seaborn
sns.barplot(x='Literacy_Rate%', y='States', data=df, palette='viridis')

# Add labels and title
plt.xlabel('Literacy_Rate%')
plt.ylabel('States')
plt.title('Literacy Rate of Each States in India')

# Display the plot
plt.show()
```

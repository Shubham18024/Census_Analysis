```python
import pandas as pd
import matplotlib.pyplot as pli

# Assuming your CSV has columns 'category' and 'value'
categories = ['Literate without educational level literacy rates (%)','below primary literacy(%)','primary literacy rates(%)','middle literacy rates(%)','matric literacy rates(%)','higher secondary literacy rates (%)','tech/notech and cert/diploma not equal to degree literacy rates(%)','Graduate & above Population literacy rates(%)']

values = [3.597818604,16.10706512,23.65051462,17.81984931,17.07569176,9.971274274,0.906306518,10.8714798]

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Northern Literacy rates (%)')

# Display the chart
plt.show()
```

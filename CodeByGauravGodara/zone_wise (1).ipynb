#  Compare the national average literacy rate with the literacy rate of  Zonal Courcils (2011)
## Aim - 
Relation Between the  Zonal Councils literacy rate and national average of literacy rate using graphs.
## Formula used -
Literacy Rate(%)= (Total literates/Population Aged 7 and above)*100
## Data Source -
The Data used for the analysis is sourced from census website (https://censusindia.gov.in/census.website/data/census-tables#). 
## Refined Data -
Here is the data source of CSV files used in this analysis [(zones_literacy_file)(https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/ZONEWISE_EDUCATION_2011.csv)]

## Zone Divided -
India is divided into 6 different zones on the basis of (North,south,east,west,central and northeast). The information regarding which state lies in which region can be found here from the official government website [https://www.mha.gov.in/sites/default/files/ZCS-CitiCharter-130710_1.pdf].
## Visualisation
Below is the code written to generate a Bar chat of literacy rates Zonal Coursils along with a comparsion to the national average literacy rate.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_zone_wise_literacy = pd.read_csv('ZONEWISE_EDUCATION_2011.csv')
df_zone_wise_literacy
df_zone_wise_literacy.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# The literacy rates of India (2011)  
literacy_rate_india_2011 = 72.98

# Load the data from the CSV file
df_zone_wise_literacy= pd.read_csv('ZONEWISE_EDUCATION_2011.csv')

# Sort data by literacy rate for better visualization
df_zone_wise_literacy= df_zone_wise_literacy.sort_values(by ='LITERACY RATES (%)')

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))
sns.barplot(y='LITERACY RATES (%)', x='ZONAL COUNCILS', color = "blue" ,data=df_zone_wise_literacy )
# Add labels and title
plt.ylabel('LITERACY RATES')
plt.xlabel('ZONAL COUNCILS')
plt.title('Literacy Rate of Each Zones in India (2011)')
plt.xticks(rotation=90)
plt.axhline(y=literacy_rate_india_2011, color='r', linestyle='--', label='India Literacy Rate')
plt.show()
```

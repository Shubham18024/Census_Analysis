# **Compare the national average literacy rate with the literacy rate of each state (1991,2001,2011)**
## Target
### Relation Between the states literacy rate and national average of literacy rate

## formula used
### Literacy rate: (Total literates/Population Aged 7 and above)*100
## Data Source
### The Data used for the analysis is sourced from census website (https://censusindia.gov.in/census.website/data/census-tables#). Here is the data source of CSV files used in this analysis [file1_1991](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/1991%20Literacy%20Rate%20and%20Growth%20Rate.xls.csv),[file2_2001](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/2001%20Literacy%20Rate%20and%20Growth%20Rate.csv),[file3_2011](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/2011%20Literacy%20Rate%20and%20Growth%20Rate.csv) 



```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census_1991 = pd.read_csv('1991 Literacy Rate and Growth Rate.xls.csv')
census_1991
census_2001 = pd.read_csv('2001 Literacy Rate and Growth Rate.csv')
census_2001
census_2011 = pd.read_csv('2011 Literacy Rate and Growth Rate.csv')
census_2011
```

```python
# Creating subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot 1991 data
sns.barplot(ax=axes[0], data=census_1991, x="Area", y="Literacy Rate 1991 (%)")
axes[0].set_title("1991 Census Data")
plt.setp(axes[0].get_xticklabels(), rotation=90, ha="right")

# Plot 2001 data
sns.barplot(ax=axes[1], data=census_2001, x="Area", y="Literacy Rate (2001) (%)")
axes[1].set_title("2001 Census Data")
plt.setp(axes[1].get_xticklabels(), rotation=90, ha="right")

# Plot 2011 data
sns.barplot(ax=axes[2], data=census_2011, x="Area", y="Literacy Rate 2011 (%)")
axes[2].set_title("2011 Census Data")
plt.setp(axes[2].get_xticklabels(), rotation=90, ha="right")

plt.tight_layout()
plt.show()
```

```python
# The literacy rates of India for each year
literacy_rate_india_1991 = 52.21 
literacy_rate_india_2001 = 64.84  
literacy_rate_india_2011 = 72.98

# Creating subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot 1991 data
sns.barplot(ax=axes[0], data=census_1991, x="Area", y="Literacy Rate 1991 (%)")
axes[0].set_title("1991 Census Data")
axes[0].axhline(y=literacy_rate_india_1991, color='r', linestyle='--', label='India Literacy Rate')
axes[0].legend()
plt.setp(axes[0].get_xticklabels(), rotation=90, ha='right')  

# Plot 2001 data
sns.barplot(ax=axes[1], data=census_2001, x="Area", y="Literacy Rate (2001) (%)")
axes[1].set_title("2001 Census Data")
axes[1].axhline(y=literacy_rate_india_2001, color='r', linestyle='--', label='India Literacy Rate')
axes[1].legend()
plt.setp(axes[1].get_xticklabels(), rotation=90, ha='right')  

# Plot 2011 data
sns.barplot(ax=axes[2], data=census_2011, x="Area", y="Literacy Rate 2011 (%)")
axes[2].set_title("2011 Census Data")
axes[2].axhline(y=literacy_rate_india_2011, color='r', linestyle='--', label='India Literacy Rate')
axes[2].legend()
plt.setp(axes[2].get_xticklabels(), rotation=90, ha='right') 

plt.tight_layout()
plt.show()
```

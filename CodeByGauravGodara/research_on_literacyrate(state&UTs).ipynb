```python
#  Compare the national average literacy rate with the literacy rate of  states and Union territories (1991,2001,2011)
## Aim - 
Relation Between the states literacy rate and national average of literacy rate using graphs
## Formula used -
Literacy rate: (Total literates/Population Aged 7 and above)*100
## Data Source -
The Data used for the analysis is sourced from census website (https://censusindia.gov.in/census.website/data/census-tables#). 
## Refined Data -
Here is the data source of CSV files used in this analysis[file1_state_1991](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/1991-STATE.csv),[file2_ut_1991](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/LITERACY1991-UT.csv),[file3_state_2001](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/LITERACY2001-STATE.csv),[file4_ut_2001](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/LITERACY2001-UT.csv),[file5_state_2011](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/LITERACY2011-STATE.csv),[file6_ut_2011](https://github.com/Shubham18024/Census_Analysis/blob/main/CSV%20files%20for%201991%2C2001%2C2011/LITERACY2011-UT.csv)
## Visualisation
Below is the code written to generate a Bar chat of literacy rates states and Union terr along with a comparsion to the national average literacy rate.
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1991_state = pd.read_csv('1991-STATE.csv')
df_1991_state
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# The literacy rates of India for each year
literacy_rate_india_1991 = 52.21 
literacy_rate_india_2001 = 64.84  
literacy_rate_india_2011 = 72.98

# Load the data from the CSV file
df = pd.read_csv('1991-STATE.csv')

# Sort data by literacy rate for better visualization
df = df.sort_values(by ='Literacy Rate 1991 (%)')

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))
sns.barplot(y='Literacy Rate 1991 (%)', x='Area', color = "blue" ,data=df )
# Add labels and title
plt.ylabel('Literacy Rate (%) ')
plt.xlabel('Area')
plt.title('Literacy Rate of Each States in India (1991)')
plt.xticks(rotation=90)
plt.axhline(y=literacy_rate_india_1991, color='r', linestyle='--', label='India Literacy Rate')
plt.show()
```

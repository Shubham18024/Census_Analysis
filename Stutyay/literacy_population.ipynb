```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census_1991 = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\CENSUS1991.csv")
census_2001 = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\2001 Data Education^J Population^J Religion^J Workers.csv")
census_2011 = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\2011 Data Education^J Population^J Religion^J Workers.csv")
census_1991
census_2001
census_2011

```

```python
columns_needed = ["Literacy_Rate%" , "Total_population"]
census_1991 = census_1991[columns_needed]
census_2001 = census_2001[columns_needed]
census_2011 = census_2011[columns_needed]
```

```python
#Creating subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
# Plot 1991 data
sns.lineplot(ax=axes[0], data=census_1991, x="Total_Population", y="Literacy_Rate%")
axes[0].set_title("1991 Census Data")

# Plot 2001 data
sns.lineplot(ax=axes[1], data=census_2001, x="Total_Population", y="Literacy_Rate%")
axes[1].set_title("2001 Census Data")

# Plot 2011 data
sns.lineplot(ax=axes[2], data=census_2011, x="Total_Population", y="Literacy_Rate%")
axes[2].set_title("2011 Census Data")

# Show the plots
plt.tight_layout()
plt.show()
```

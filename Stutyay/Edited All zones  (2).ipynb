# Analysis of Literacy Rate by Gender with Respect to Rural and Urban Divide by Zonal Division

## Aim
The objective of this analysis is to visualize the literacy rates of males and females with the urban and rural divide among the six different zones (North, South, East, West, Central, and Northeast). The information regarding which state lies in which region can be found on the official government website: [Zones](https://www.mha.gov.in/sites/default/files/ZCS-CitiCharter-130710_1.pdf).

## Formulas
The formulas used for calculating the literacy rates of males and females with respect to the urban and rural divide are as follows:
$$
\text{Rural Literacy Rate} = \left( \frac{\text{Total Rural Literates (age 7+)}}{\text{Total Rural Population (age 7+)}} \right) \times 100
$$

$$
\text{Rural Male Literacy Rate} = \left( \frac{\text{Rural Male Literates (age 7+)}}{\text{Total Rural Male Population (age 7+)}} \right) \times 100
$$

$$
\text{Rural Female Literacy Rate} = \left( \frac{\text{Rural Female Literates (age 7+)}}{\text{Total Rural Female Population (age 7+)}} \right) \times 100
$$

$$
\text{Urban Literacy Rate} = \left( \frac{\text{Total Urban Literates (age 7+)}}{\text{Total Urban Population (age 7+)}} \right) \times 100
$$

$$
\text{Urban Male Literacy Rate} = \left( \frac{\text{Urban Male Literates (age 7+)}}{\text{Total Urban Male Population (age 7+)}} \right) \times 100
$$

$$
\text{Urban Female Literacy Rate} = \left( \frac{\text{Urban Female Literates (age 7+)}}{\text{Total Urban Female Population (age 7+)}} \right) \times 100
$$

$$
\text{Gender Gap Rural} = \text{Rural Male Literacy Rate} - \text{Rural Female Literacy Rate}
$$

$$
\text{Gender Gap Urban} = \text{Urban Male Literacy Rate} - \text{Urban Female Literacy Rate}
$$

## Visualization
Below is the code written to generate a grouped bar chart of the literacy rates of males and females with the rural and urban divide among six different zones, along with a comparison to the national average literacy rate.

```python
# Importing the Necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the CSV file 
df = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\All zones combined.csv")
df
```

```python
#Selecting the relevant columns from the df
rural_data = df[["Zone", "Rural_Male_LiteracyRate(age 7+)", "Rural_Female_LiteracyRate(age 7 +)"]]

# Set up the plot
fig, ax = plt.subplots(figsize=(18, 12))

# Set the width of each bar and the positions of the bar 
bar_width = 0.35
r1 = range(len(rural_data))
r2 = [x + bar_width for x in r1]

# Create the bars
male_bars = ax.bar(r1, rural_data["Rural_Male_LiteracyRate(age 7+)"], color="blue", width=bar_width, label="Males")
female_bars = ax.bar(r2, rural_data["Rural_Female_LiteracyRate(age 7 +)"], color="orange", width=bar_width, label="Females")

# Add horizontal line for national average
national_avg = 67.76
national_avg_line = ax.axhline(y=national_avg, color="r", linestyle="--", linewidth=2, label=f'Rural Avg Literacy: {national_avg}%')

# Customize the plot
ax.set_ylabel("Literacy Rates (%)", fontsize=25, fontweight='bold', fontname="Times New Roman")
ax.set_xticks([r + bar_width/2 for r in range(len(rural_data))])
ax.set_xticklabels(rural_data["Zone"], rotation=40, ha='right', fontsize=14, fontweight='bold', fontname="Times New Roman")
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x', labelsize=18)

# Adjust y-axis scale
ax.set_ylim(0, 100)
ax.set_yticks(np.arange(0, 101, 20))

# Add percentage values on top of each bar
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=14)

autolabel(male_bars)
autolabel(female_bars)


# Adjust legend to avoid overlap and place it outside the plot
ax.legend([male_bars, female_bars, national_avg_line], 
          ['Males', 'Females', f'Rural Avg literacy: {national_avg}%'], 
          loc="upper left", bbox_to_anchor=(1,1), fontsize=14)

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(right=0.85)

# Save the Rural Literacy Rates image
plt.savefig('Rural_Literacy_Rates.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census_2001=pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\Copy of 2001_Literacy_Rate_and_Growth_Rate(1).csv")
census_2011=pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\Copy of 2011_Literacy_Rate_and_Growth_Rate(1).csv")
census_2001.columns
census_2011.columns
```

```python
columns_needed = ["Change_in_LiteracyRate" , "GrowthRate"]
census_2001 = census_2001[columns_needed]
census_2011 = census_2011[columns_needed]
```

```python
#Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(12,4))

# Plot 2001 data
sns.lineplot(ax=axes[0], data=census_2001, y= 'Change in Literacy Rate (1991 & 2001)(%)', x='Growth Rate (1991 & 2001) (%)')
axes[0].set_title("2001 Census Data")


# Plot 2011 data
sns.lineplot(ax=axes[1], data=census_2011, y='Change in Literacy Rate (2001 & 2011) (%)', x='Growth Rate (2001 & 2011) (%)')
axes[1].set_title("2011 Census Data")

# Show the plots
plt.tight_layout()
plt.show()
```

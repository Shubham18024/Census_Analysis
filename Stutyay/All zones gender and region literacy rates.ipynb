#Analysis of Literacy Rate by Gender with respect to Rural and Urban divide by Zonal division


##Aim
The objective of this analysis to visualize the literacy rate of males and females with the urban and rural divide amongst the 6 different zones(North,south,east,west,central and northeast). The information regarding which state lies in which region can be found here from the official government website: [Zones](https://www.mha.gov.in/sites/default/files/ZCS-CitiCharter-130710_1.pdf)


##Formula
The formulas used for calculating the literacy rates of males and females with respect to Urban and rural divide are as follows:
                     Rural Literacy Rate: (Total Rural literates(age 7+)/Total Rural Population (age 7+))100
                     Rural Male literacy rate : (Rural male literates(age 7+)/Total Rural male Population(age 7+))100
                     Rural Female literacy rate : (Rural female literates(age 7+)/Total Rural female Population(age 7+))100
                     Urban Literacy rate: (Total Urban literates(age 7+)/Total Urban Population (age 7+))100
                     Urban Female literacy rate:(Urban female literates(age 7+)/Total Urban female Population(age 7+))100
                     Urban Male literacy rate:(Urban male literates(age 7+)/Urban Rural male Population(age 7+))100
                     Gender Gap Rural: Rural Male literacy rate -  Rural Female literacy rate 
                     Gender Gap Urban: Urban Male literacy rate - Urban Female literacy rate

##Visualization
Below is the code written to generate a grouped bar chat of literacy rate of males and females with rural and urban divide amongst 6 different zone along with the comparison to the national average literacy rate.

```python
# Importing the Necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the CSV file 
df = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\All zones combined.csv")
df
```

```python
# Selecting the relevant columns from the df
rural_data = df[["Zone", "Rural_Male_LiteracyRate(age 7+)", "Rural_Female_LiteracyRate(age 7 +)"]]

# Set up the plot
fig, ax = plt.subplots(figsize=(18, 12))

# Set the width of each bar and the positions of the bar 
bar_width = 0.35
r1 = range(len(rural_data))
r2 = [x + bar_width for x in r1]

# Create the bars
male_bars = ax.bar(r1, rural_data["Rural_Male_LiteracyRate(age 7+)"], color="blue", width=bar_width, label="Males")
female_bars = ax.bar(r2, rural_data["Rural_Female_LiteracyRate(age 7 +)"], color="orange", width=bar_width, label="Females")

# Add horizontal line for national average
national_avg = 67.76
national_avg_line = ax.axhline(y=national_avg, color="r", linestyle="--", linewidth=2, label=f'National Avg of rural Literacy: {national_avg}%')

# Customize the plot
ax.set_ylabel("Literacy Rates (%)", fontsize=25, fontweight='bold', fontname="Times New Roman")
ax.set_title("Rural Literacy Rates by Gender", fontsize=38, fontweight='bold', fontname="Times New Roman", pad=20)
ax.set_xticks([r + bar_width/2 for r in range(len(rural_data))])
ax.set_xticklabels(rural_data["Zone"], rotation=40, ha='right', fontsize=14, fontweight='bold', fontname="Times New Roman")
ax.tick_params(axis='y', labelsize=16)
ax.tick_params(axis='x', labelsize=18)

# Adjust y-axis scale
ax.set_ylim(0, 100)
ax.set_yticks(np.arange(0, 101, 20))

# Add percentage values on top of each bar
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=14)

autolabel(male_bars)
autolabel(female_bars)

# Adjust legend to avoid overlap
ax.legend([male_bars, female_bars, national_avg_line], 
          ['Males', 'Females', f'National Avg of rural Literacy: {national_avg}%'], 
          loc="upper right", bbox_to_anchor=(1,1), fontsize=14)

# Adjust layout
fig.tight_layout()

# Show the plot
plt.show()
```

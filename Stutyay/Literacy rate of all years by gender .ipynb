```python
 #Importing the Libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df=pd.read_excel(r"C:\Users\stuti\OneDrive\Desktop\temp_1\Copy of Literacy_Rates_All_Years_New(1).xlsx" , engine="openpyxl")
df
```

```python
#Stating the columns
df.columns
```

```python
# Set the style for seaborn
sns.set(style='whitegrid')

# Create figure and axes objects
fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(data=df, x="CENSUS YEARS", y="MALES(Literacy rate)", label="Males", 
             color='green', linewidth=3, linestyle='--', ax=ax)
sns.lineplot(data=df, x="CENSUS YEARS", y="FEMALES(Literacy rate)", label="Females", 
             color='orange', linewidth=3, linestyle=':', ax=ax)
sns.lineplot(data=df, x="CENSUS YEARS", y="ALL PERSONS(Literacy rate)", label="All Persons Literacy Rate", 
             color='blue', linewidth=3, linestyle='-', ax=ax)
sns.lineplot(data=df, x="CENSUS YEARS", y="MALE FEMALE GAP", label="Gender Gap", 
             color='red', linewidth=3, linestyle='-.', ax=ax)

# Add title and labels
plt.ylabel('Rate',fontsize=25, fontweight='bold', fontname="Times New Roman")

# Add title and labels
plt.ylabel('Crude Literacy Rates (%)', fontsize=14, fontweight='bold')

# Set x-ticks to include all years from 1901 to 2011 with a 10-year gap
plt.xticks(ticks=range(1901, 2012, 10))

# Remove x-axis label
plt.xlabel('') 

# Customize and position the legend
legend = plt.legend(loc='upper right', bbox_to_anchor=(0.98, 0.98), frameon=True, framealpha=1, edgecolor='black')

# Extend y-axis to 100
plt.ylim(0, 100)

# Display legend
plt.legend()


# Save the image of this graph
plt.savefig("Literacy of all years by gender and total.pdf", dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```

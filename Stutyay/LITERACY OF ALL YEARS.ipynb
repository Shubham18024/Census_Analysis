```python
#Importing the Libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the Dataset
df = pd.read_excel(r"C:\Users\stuti\OneDrive\Desktop\temp_1\Literacy Rates All Years New.xlsx", engine="openpyxl")
df
```

```python
df.columns
```

```python
# Set the style for seaborn
sns.set(style='whitegrid')

# Create line plots
plt.figure(figsize=(12, 8))

# Plot each line on the same graph
sns.lineplot(data=df, x="CENSUS YEARS", y="FEMALES(Literacy rate)", label="Male Literacy Rate")
sns.lineplot(data=df, x="CENSUS YEARS", y="MALES(Literacy rate)", label="Female Literacy Rate")
sns.lineplot(data=df, x="CENSUS YEARS", y="ALL PERSONS(Literacy rate)", label="All Persons Literacy Rate")
sns.lineplot(data=df, x="CENSUS YEARS", y="MALE FEMALE GAP", label="Male-Female Gap")

# Add title and labels
plt.title('Literacy Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Rate')

# Set x-ticks to include all years from 1901 to 2011 with a 10-year gap
plt.xticks(ticks=range(1901, 2012, 10))

# Display legend
plt.legend()

# Show plot
plt.show()
```

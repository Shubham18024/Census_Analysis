```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census_1991 = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\1991_Census.csv")
census_2001 = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\2001_Census.csv")
census_2011 = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\temp_1\2011_Census.csv")
census_1991.columns
```

```python
columns_needed = ["LiteracyRate" , "GrowthRate"]
census_1991 = census_1991[columns_needed]
census_2001 = census_2001[columns_needed]
census_2011 = census_2011[columns_needed]
```

```python
#Creating subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
# Plot 1991 data
sns.lineplot(ax=axes[0], data=census_1991, x="X..1981 to 1991", y="Literacy_rates")
axes[0].set_title("1991 Census Data")

# Plot 2001 data
sns.lineplot(ax=axes[1], data=census_2001, x="X..1991 to 2001", y="Literacy_Rates")
axes[1].set_title("2001 Census Data")

# Plot 2011 data
sns.lineplot(ax=axes[2], data=census_2011, x="X..2001 to 2011", y="Literacy_rates")
axes[2].set_title("2011 Census Data")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#Creating subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
# Plot 1991 data
sns.scatterplot(ax=axes[0], data=census_1991, x="X..1981 to 1991", y="Literacy_rates")
axes[0].set_title("1991 Census Data")

# Plot 2001 data
sns.scatterplot(ax=axes[1], data=census_2001, x="X..1991 to 2001", y="Literacy_Rates")
axes[1].set_title("2001 Census Data")

# Plot 2011 data
sns.scatterplot(ax=axes[2], data=census_2011, x="X..2001 to 2011", y="Literacy_rates")
axes[2].set_title("2011 Census Data")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#Creating subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
# Plot 1991 data
sns.barplot(ax=axes[0], data=census_1991, x="X..1981 to 1991", y="Literacy_rates")
axes[0].set_title("1991 Census Data")
plt.setp(axes[0].get_xticklabels(), rotation=90, ha="right")

# Plot 2001 data
sns.barplot(ax=axes[1], data=census_2001, x="X..1991 to 2001", y="Literacy_Rates")
axes[1].set_title("2001 Census Data")

# Plot 2011 data
sns.barplot(ax=axes[2], data=census_2011, x="X..2001 to 2011", y="Literacy_rates")
axes[2].set_title("2011 Census Data")
plt.setp(axes[2].get_xticklabels(), rotation=90, ha="right")

# Show the plots
plt.tight_layout()
plt.show()
```

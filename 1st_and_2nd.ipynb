```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
var = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\india-districts-census-2011.csv")
var
education_level=["Below_Primary_Education", "Primary_Education" , "Middle_Education" , "Secondary_Education" , "Higher_Education" , "Graduate_Education" , "Other_Education"]
Total_Education = var[education_level].sum()
# Scatter plot for Urban Households vs Total Education
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Total_Education", y="Urban_Households", data=var)
plt.title("Scatter Plot of Urban Households vs Total Education")
plt.xlabel("Total Education")
plt.ylabel("Urban Households")
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot for Rural Households vs Total Education
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Total_Education", y="Rural_Households", data=var)
plt.title("Scatter Plot of Rural Households vs Total Education")
plt.xlabel("Total Education")
plt.ylabel("Rural Households")
plt.xticks(rotation=90)
plt.show()
```

```python
# Zoomed Scatter Plot for Urban Households
urban_range = (var["Urban_Households"] >= 9000) & (var["Urban_Households"] <= 30000)
var_zoomed_urban = var[urban_range]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Total_Education", y="Urban_Households", data=var_zoomed_urban)
plt.title("Zoomed Scatter Plot for Urban Households vs Total Education")
plt.xlabel("Total Education")
plt.ylabel("Urban Households")
plt.xticks(rotation=90)
plt.show()
```

```python
# Zoomed Scatter Plot for Rural Households
rural_range = (var["Rural_Households"] >= 10000) & (var["Rural_Households"] <=40000 )
var_zoomed_rural = var[rural_range]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Total_Education", y="Rural_Households", data=var_zoomed_rural)
plt.title("Zoomed Scatter Plot for Rural Households vs Total Education")
plt.xlabel("Total Education")
plt.ylabel("Rural Households")
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
var = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\india-districts-census-2011.csv")
var
```

```python
# Scatter plot for Urban Households vs Total Population
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Population", y="Urban_Households", data=var)
plt.title("Scatter Plot of Urban Households vs Total Population")
plt.xlabel("Total Population")
plt.ylabel("Urban Households")
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot for Rural Households vs Total Population
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Population", y="Rural_Households", data=var)
plt.title("Scatter Plot of Rural Households vs Total Population")
plt.xlabel("Total Population")
plt.ylabel("Rural Households")
plt.xticks(rotation=90)
plt.show()
```

```python
# Zoomed Scatter Plot for Urban Households
urban_range = (var["Urban_Households"] >= 1000) & (var["Urban_Households"] <= 30000)
var_zoomed_urban = var[urban_range]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Population", y="Urban_Households", data=var_zoomed_urban)
plt.title("Zoomed Scatter Plot for Urban Households vs Total Population")
plt.xlabel("Total Population")
plt.ylabel("Urban Households")
plt.xticks(rotation=90)
plt.show()
```

```python
# Zoomed Scatter Plot for Rural Households
rural_range = (var["Rural_Households"] >= 5000) & (var["Rural_Households"] <=10000 )
var_zoomed_rural = var[rural_range]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="Population", y="Rural_Households", data=var_zoomed_rural)
plt.title("Zoomed Scatter Plot for Rural Households vs Total Population")
plt.xlabel("Total Population")
plt.ylabel("Rural Households")
plt.xticks(rotation=90)
plt.show()
```

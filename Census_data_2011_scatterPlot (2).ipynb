```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
var = pd.read_csv(r"C:\Users\stuti\OneDrive\Desktop\india-districts-census-2011.csv")
var



```

```python
#Scatter PLot for Rural Household


#cluttered_region = (var["Rural_Households"] >= 0.0) & (var["Rural_Households"] <= 1.6)


#var_zoomed = var[cluttered_region]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="State name", y="Rural_Households", data=var)

plt.title("Scatter Plot of Rural Household")
plt.xticks(rotation=90)

plt.show()
```

```python
#Scatter PLot for Urban Household
plt.figure(figsize=(15, 8))
sns.scatterplot(x="State name" ,
               y="Urban_Households" , data=var)
plt.title("Scattered Plot of Urban Household")
plt.xticks(rotation=90)

plt.show()
```

```python
#Zoomed Scatter Plot for Rural Household


cluttered_region = (var["Rural_Households"] >=8000) & (var["Rural_Households"] <=90000 )


var_zoomed = var[cluttered_region]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="State name", y="Rural_Households", data=var_zoomed)

plt.title("Zoomed Scattered Plot for Rural Household")
plt.xticks(rotation=90)

plt.show()
```

```python
#Zoomed Scatter Plot for Urban Household

cluttered_region = (var["Urban_Households"] >= 1000) & (var["Urban_Households"] <= 30000)


var_zoomed = var[cluttered_region]
plt.figure(figsize=(15, 8))
sns.scatterplot(x="State name", y="Urban_Households", data=var_zoomed)

plt.title("Zoomed Scatter Plot for Urban Household")
plt.xticks(rotation=90)

plt.show()
```
